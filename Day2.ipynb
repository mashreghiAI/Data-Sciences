```python
# Q2: Consider CRM dataset and answer these questions:
crm_data = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 5],
    'marital_status': ['married', 'single', 'married', 'single', 'married'],
    'income': [50000, 40000, 60000, 45000, 70000]
})
```

```python
# a: Extract data of married customers with income above average.
average_income = crm_data['income'].mean()
married_above_average = crm_data[(crm_data['marital_status'] == 'married') & (crm_data['income'] > average_income)]
print(married_above_average)
```

```python
# b: What percentag of female customers are below 26?
crm_data = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 5],
    'gender': ['female', 'male', 'female', 'female', 'male'],
    'age': [22, 30, 24, 25, 28]
})
female_customers = crm_data[crm_data['gender'] == 'female']
below_26 = female_customers[female_customers['age'] < 26]
percentage_below_26 = (len(below_26) / len(female_customers)) * 100
print(f"Percentage of female customers below 26 years old: {percentage_below_26:.2f}%")
```

```python
# c: What percentage of female customers above 30 are employed?
crm_data = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 5],
    'gender': ['female', 'male', 'female', 'female', 'male'],
    'age': [22, 35, 32, 45, 28],
    'employment_status': ['employed', 'unemployed', 'employed', 'employed', 'unemployed']
})
female_above_30 = crm_data[(crm_data['gender'] == 'female') & (crm_data['age'] > 30)]
employed_females = female_above_30[female_above_30['employment_status'] == 'employed']
percentage_employed_females = (len(employed_females) / len(female_above_30)) * 100
print(f"Percentage of female customers above 30 years old who are employed: {percentage_employed_females:.2f}%")
```

```python
# f: What percentage of married customers have health insurance?
crm_data = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 5],
    'marital_status': ['married', 'single', 'married', 'married', 'single'],
    'health_insurance': [True, False, True, False, True]
})
married_customers = crm_data[crm_data['marital_status'] == 'married']
married_with_health_insurance = married_customers[married_customers['health_insurance'] == True]
percentage_married_with_health_insurance = (len(married_with_health_insurance) / len(married_customers)) * 100
print(f"Percentage of married customers who have health insurance: {percentage_married_with_health_insurance:.2f}%")
```
